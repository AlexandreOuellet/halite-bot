```python
import matplotlib.pyplot as plt
#v58
weights = [0.415212647528, 1.56730222479, 0.615640117132, 1.36098773839, 0.894544218321, 0.297630730634, -0.162094705167, 0.342260804162, -0.361209049613]
#v8
#weights =[2.35921429811,-0.940531037825, -0.591145702795, 0.282752707451, -0.223289091834, 1.99096211367, 0.19317830957, -0.0350884461805, 0.639219766106]
friendlyWeight = weights[1]
enemyWeight = weights[2]
neutralWeight = weights[3]
fullyEnemyDocked = weights[7]
fullyFriendlyDocked = weights[8]
```

```python
# 
distance = []
for i in range(0, 100, 10):
    value = 1 - i*0.01
    distance.append(weights[0] * value)
plt.plot(distance)
```

```python
friendly = []
for i in range(0, 100, 10):
    value = i*0.10
    friendly.append(weights[1] * value)
friendly
plt.plot(friendly)
```

```python
# 
enemy = []
for i in range(0, 100, 10):
    value = i*0.10
    enemy.append(weights[2] * value)
enemy
plt.plot(enemy)
```

```python
distanceSquared = []
for i in range(0, 100, 10):
    value = i*0.01
    value = 1 - (value * value)
    distanceSquared.append(weights[4] * value)
distanceSquared
plt.plot(distanceSquared)
```

```python
health = []
for i in range(0, 100, 10):
    value = i*0.01
    health.append(weights[5] * value)
health
plt.plot(health)
```

```python
neutralCapacity = []
for i in range(0, 100, 10):
    value = i*0.01
    neutralCapacity.append(weights[6] * value)
neutralCapacity
plt.plot(neutralCapacity)
```

```python
enemeyDockedShip = []
for i in range(0, 100, 10):
    value = i*0.01
    enemeyDockedShip.append(weights[7] * value)
enemeyDockedShip
plt.plot(enemeyDockedShip)
```

```python
friendlyDockedShip = []
for i in range(0, 100, 10):
    value = i*0.01
    friendlyDockedShip.append(weights[8] * value)
friendlyDockedShip
plt.plot(friendlyDockedShip)
```

```python
x=range(0,10)
plt.stackplot(x,distance, friendly, enemy, distanceSquared, health, neutralCapacity, enemeyDockedShip, friendlyDockedShip, labels=["distance", "friendly", "enemy", "distanceSquared", "health", "neutralCapacity", "enemeyDockedShip", "friendlyDockedShip"])
```

```python
fig, ax = plt.subplots()
ax.stackplot(x,distance, friendly, enemy, distanceSquared, health, neutralCapacity, enemeyDockedShip, friendlyDockedShip, labels=["distance", "friendly", "enemy", "distanceSquared", "health", "neutralCapacity", "enemeyDockedShip", "friendlyDockedShip"])
ax.legend(loc=0)
plt.show()
```
